```python
from starkiller import sat_killer
import cv2
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
image = cv2.imread('thumb.png',0)
```

```python
plt.figure()
plt.imshow(image,origin='lower')
```

```python
sat = sat_killer(1,1,run=False)
sat.image = image
sat._set_threshold(3)
print(sat.threshold)
```

```python
sat._dilate()
sat._edges()
sat._lines()
sat._match_lines()
sat._find_center()
sat.make_mask()
sat._detected()
```

```python
sat.plot_lines()
plt.savefig('thumbnail_test.png')
```
